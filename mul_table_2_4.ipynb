```python
def mul(num1,num2):
    #1 0000 = 0011,10 0000 = 0110,100 0000 = 1100,1000 0000 = 1011
    cal_num = [3,6,12,11]
    out = 0
    for i in range(4):
        if ((num1 >> i) & 1) == 1:
            out ^= (num2 << i)
    for j in range(4,8):
        if((out >> j) & 1) == 1:
            out ^= 1 << j
            out ^= cal_num[j-4]
    #print(num1,num2,out)
    return out

list = [[0 for i in range(16)] for j in range(16)]
for i in range(16):
    print(i,"&")
    for j in range(16):
        list[i][j] = mul(i,j)
        if j == 15:
            print(format(mul(i,j), 'x'),"\\\\")
        else:
            print(format(mul(i,j), 'x'),"&")
    print("\n")
    #print(list[i])
```
